```python
from pyspark.sql.functions import col

STORAGE_PATH = "abfss://fraud-sentinel@giftmapote2ete.dfs.core.windows.net/raw/transactions/"
# 1. Define the stream again (ensure STORAGE_PATH is correct)
# We use Append mode because we are simply listing new records
inspection_stream = (spark.readStream
                     .format("cloudFiles")
                     .option("cloudFiles.format", "json")
                     .option("cloudFiles.schemaLocation", f"{STORAGE_PATH}_schema_peeker")
                     .load(STORAGE_PATH))

# 2. Display the live data
# This will create a dashboard-style view that updates every time a new batch lands.
# 2. Display using 'Append' logic (implicitly handled by display() for simple selects)
# We remove 'orderBy' because global sorting requires 'Complete' mode, 
# which triggers the error you saw on non-aggregated streams
display(inspection_stream.select(
    "transaction_id", 
    "customer_id", 
    "amount", 
    "location", 
    "timestamp"
))
```
